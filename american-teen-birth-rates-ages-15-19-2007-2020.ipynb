## A Sum of Teen Birth Rates in Ages 15-19 Across the US (2007-2020)

```r
if (!requireNamespace("dplyr", quietly = TRUE)) {
  install.packages("dplyr")
}
library(dplyr)
```

```r
sum_birth_rates <- data %>%
  group_by(State) %>%
  summarise(Total_Birth_Rate = sum(Birth.Rate)) %>%
  arrange(desc(Total_Birth_Rate))

# Print the result
print(sum_birth_rates)
```

The result shows the total birth rates summarized for each state, arranged in descending order. Here are some observations that I have found:

1. **Texas has the highest total birth rate:** Texas leads with a total birth rate of 234,899. This means that, based on the provided data, Texas had the highest number of births among the states.

2. **Southern states dominate the top positions:** States like Georgia, Kentucky, Missouri, Mississippi, and Tennessee follow Texas in the top positions. This suggests a higher concentration of births in the southern region of the United States.

3. **Variation in birth rates among states:** The total birth rates vary significantly among states, indicating differences in population size, demographics, and possibly other factors influencing birth rates.


## A Sum of American Teen Birth Rates in Ages 15-19 Throughout the Years (2007-2020)

```r
sum_birth_rates <- data %>%
  group_by(Year) %>%
  summarise(Total_Birth_Rate = sum(Birth.Rate)) %>%
  arrange(desc(Total_Birth_Rate))

# Print the result
print(sum_birth_rates)
```

The provided results show the total birth rates summarized for each year, arranged in descending order. Here are some observations that I found:

1. **Peak years for total birth rates:** The years 2007 and 2008 had the highest total birth rates, with 147,773 and 145,571 births, respectively. These were peak years for the total number of births among the years covered by the data.

2. **Decline in total birth rates:** There is a general trend of declining total birth rates over the years. The number of births decreased from 141,101 in 2003 to 68,095 in 2020.

4. **Insights for further investigation:** The declining trend may raise questions about factors influencing birth rates, such as economic conditions, social trends, and policy changes. Further investigation could explore these factors and their impact on birth rates over time.
